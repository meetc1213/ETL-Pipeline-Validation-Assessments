```python
# load the data using pandas
import pandas as pd
import numpy as np
pd.set_option('display.max_colwidth', None)
df1 = pd.read_csv("QuestionnaireQuestions-202209061907.csv")
df2 = pd.read_csv("SubmittedQuestionResponses-202209061913.csv")
pd.set_option('display.max_rows', 10)
from IPython.display import display


new_values = [] # holds new response ids which I recreate from the response_set_id
for row in df2.itertuples(index=False):
    splitted = row.response_set_id.split('.')
    question_id = splitted[0][2:7]
    if row.lang == 'en':
        lang_code = '.e11.'
    else:
        lang_code = '.e28.'
    response_id = 'i'+question_id+lang_code+splitted[-1] # I take the option code (r0,r1,..) code which is given in response_set_id
    new_values.append(response_id)

df2['n_response_id'] = new_values
df2 = df2[['response_set_id','n_response_id', 'response_id', 'is_correct',
                                 'response', 'response_text', 'score', 'is_nr',
                                    'question_label','is_questionnaire',"is_excluded","is_invalid","is_operational","is_sample"]]

merged = df1.merge(df2, left_on='response_id', right_on='n_response_id', how='inner')
merged = merged[['id', 'response_id_x', 'response_set_id','label',
       'is_correct', 'response', 'response_text', 'score', 'is_nr']]
# merged[merged['label']=='Question 15'] # less than total in df1 because not everything has a response
merged = merged[merged['label']=='Question 5']

print(merged.to_string())
```
