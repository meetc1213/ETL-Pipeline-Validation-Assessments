```python
# load the data using pandas
import pandas as pd
import numpy as np
import json

pd.set_option('display.max_colwidth', None)
df = pd.read_csv("test_attempt_question_responses-202209061938.csv")


rows = []
for row in df.itertuples(index=False):
    response = json.loads(row.response_raw)['3']

    # Advanced check: check if each of the targets has atleast a non empty content. If so, it is responded.
    # Actually gave 20 rows where response['isResponded'] == False eventhough there are responses
    targets = response.get('targets')
    responded = False # this is the flag which checks if atleast one content is not empty

    for target in targets:
        if target['contents'] != []:
            responded = True
            break

    if not (responded and response['isResponded']):
        # print('Yoo misaligned')
        rows.append({
        "id":row.id,
        "test_question_id": row.test_question_id,
        "is_nr": row.is_nr,
        "isResponded": response['isResponded'],
    })



df_misaligned = pd.DataFrame(rows)

df_misaligned
```
